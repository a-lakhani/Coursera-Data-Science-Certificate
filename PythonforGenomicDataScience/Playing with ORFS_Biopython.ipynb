```python
import os
import sys
from pprint import pprint

# Seq is a class of immutable string-like objects with additional methods like transcribe & translate. 
from Bio.Seq import Seq
from Bio import SeqIO

# NCBIWWW.qblast queries the NCBI BLAST API and returns an XML file.
# NCBIXML.read (for single BLAST query) and NCBIXML.parse (for multiple BLAST queries)
# parse XML files returned by NCBIWWW.qblast into special Blast Record objects, 
# which are described in the Bio.Blast.Record module.
from Bio.Blast import NCBIWWW, NCBIXML
```

```python
path_to_fasta_file = '/home/alakhani/Coursera-Projects/PythonforGenomicDataScience/FASTA_Files/dna.example_copy.fasta'
```

```python
# with automatically closes file before leaving statement. 
# A bonus is being able to retry things without having to file.seek(0) every time
with open(path_to_fasta_file,'r') as fasta_file:
    pprint([i for i in fasta_file])
```

```python
# FASTA record descriptions begin with '>', 
# which allows us to count the total number of records in the file.
with open(path_to_fasta_file,'r') as fasta_file:
    fasta_records = 0
    for line in fasta_file:
        if line[0] == '>':
            fasta_records += 1
        
print(fasta_records)
```
